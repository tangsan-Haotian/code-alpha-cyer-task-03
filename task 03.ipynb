```python
import matplotlib.pyplot as plt
import time
import random

def simulate_device_scan():
    sample_devices = ["aa:bb:cc:dd:ee:ff", "11:22:33:44:55:66", "77:88:99:00:aa:bb", "cc:dd:ee:ff:gg:hh"]
    detected_devices = set(random.choices(sample_devices, k=3))
    return detected_devices

def check_unknown_devices(detected_devices, known_devices):
    unknown_devices = detected_devices - known_devices
    if unknown_devices:
        print("Unknown RFID/Signals detected! Possible unauthorized reader:")
        for device in unknown_devices:
            print(f" - Device MAC: {device}")
    else:
        print("No unknown devices detected.")

def plot_devices(detected_devices):
    plt.figure(figsize=(8, 6))
    plt.title("Detected Wireless Devices")
    plt.bar(list(range(len(detected_devices))), [1]*len(detected_devices), color="blue")
    plt.xticks(range(len(detected_devices)), detected_devices, rotation=45)
    plt.show()

def main():
    known_devices = {"aa:bb:cc:dd:ee:ff"}
    print("Starting RFID security detection simulation...")
    while True:
        detected_devices = simulate_device_scan()
        check_unknown_devices(detected_devices, known_devices)
        plot_devices(detected_devices)
        time.sleep(5)

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import time
import random

def simulate_hacking_attempts():
    hacking_attempts = {
        "Attempt_1": random.randint(1, 10),
        "Attempt_2": random.randint(1, 10),
        "Attempt_3": random.randint(1, 10),
        "Attempt_4": random.randint(1, 10),
        "Attempt_5": random.randint(1, 10)
    }
    return hacking_attempts

def display_hacking_attempts(attempts):
    plt.figure(figsize=(8, 6))
    plt.title("Hacking Attempts Detected")
    plt.bar(attempts.keys(), attempts.values(), color="red")
    plt.xlabel("Attempt ID")
    plt.ylabel("Attempt Count")
    plt.show()

def main():
    print("Starting simulated hacking attempt detection...")
    while True:
        attempts = simulate_hacking_attempts()
        print("Hacking Attempts:", attempts)
        display_hacking_attempts(attempts)
        time.sleep(5)

if __name__ == "__main__":
    main()
```
